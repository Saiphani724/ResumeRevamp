```python
import os
import json
import openai
import random
```

```python
## PASTE key from .env 
model = 'gpt-3.5-turbo-0613'
openai.api_key = os.getenv('OPENAI_API_KEY')
```

```python
dir_path = './resume_txt'
files = os.listdir(dir_path)
file_name = random.choice(files)
print(file_name)
with open(os.path.join(dir_path, file_name), 'r') as file:
    resume_txt = file.read()
```

```python
prompt = f"""

Please provide me the following in JSON format.

personal_details which should Include  name, location, phone, email, linkedin, and github.

education should be categorized as school/university , duration_years, gpa, courses, etc.

experience should be categorized as company_name, duration and summary.

Keep summary as it as and don't change anything.

projects also similar to experience.

technial skills, it has different sections like languages, frameworks, tools, databases. 

extra_curriculars if any sort them into summary containg the exact information.

if any information provided in the resume doesn't fall under one of the above categories, 
include these sections in others, include each section under name of section

do not include any information twice.

Here is the resume. You will have parse this into above mentioned format.
{resume_txt}

"""
```

```python
def get_messages(prompt):
  messages = [
    {
      "role": "system",
      "content": "You are resume parser. Your task is to parse given resume into given format without excluding any details from the resume. Try to find a section where the info fits, else dump it seperately under others section."
    },
    {
    "role": "user",
    "content": prompt,
    }
  ]
  return messages

def get_completion(messages):
        response = openai.ChatCompletion.create(
            model=model,
            messages=messages,
            temperature=0
        )
        return response


import json
(json.dumps(prompt))

result = get_completion(get_messages(prompt))
print(result.choices[0]['message']['content'])
```
