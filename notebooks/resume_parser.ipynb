```python
def get_messages(prompt):
  messages = [
    {
      "role": "system",
      "content": "You are resume parser. Your task is to parse given resume into given format without excluding any details from the resume. Try to find a section where the info fits, else dump it seperately under others section."
    },
    {
    "role": "user",
    "content": prompt,
    }
  ]
  return messages

def get_completion(messages):
        response = openai.ChatCompletion.create(
            model=model,
            messages=messages,
            temperature=0
        )
        return response


(json.dumps(prompt))

result = get_completion(get_messages(prompt))
print(result.choices[0]['message']['content'])
```

```python
def extract_resume_section(fns, messages):
    #print(messages)
    completion = openai.ChatCompletion.create(
        model="gpt-4-0613",
        messages=messages,
       
        functions=fns,
        function_call={
           "name": f"revamp_resume"
        }
    )
    return completion.choices[0].message
```
