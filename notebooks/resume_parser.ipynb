```python
import os
import json
import openai
import random
```

```python
## PASTE key from .env 
model = 'gpt-3.5-turbo-0613'
openai.api_key = os.getenv('OPENAI_API_KEY')
```

```python
dir_path = '../db/resume_txt'
files = os.listdir(dir_path)
file_name = random.choice(files)
print(file_name)
with open(os.path.join(dir_path, file_name), 'r', encoding='utf-8') as file:
    resume_txt = file.read()
```

```python
prompt = f"""

Please provide me the following in JSON format:

- "personal_details" which should include: name, location, phone, email, LinkedIn, and GitHub.
- "education" should be categorized as school/university, duration_years, GPA, courses.
- "experience" should be categorized as company_name, duration, and summary.
- Keep the "summary" as it is, and don't change anything.
- "projects" should be organized similarly to "experience."
- "technical skills" has different sections like languages, frameworks, tools, databases.
- Add "course_work" section containing courses if only the courses are not mentioned in "education"
- "extra_curriculars," if any, should be sorted into "summary" containing the exact information.
- If any information provided in the resume doesn't fall under one of the above categories,
  include these sections in "others." Include each section under the name of the section.

Do not include any information twice.

Here is the resume. You will have to parse this into the above-mentioned format:
{resume_txt}

"""
```

```python
basic_prompt = f"""

Please provide me the following in JSON format:

Extract details from the resume 

Here is the resume. You will have to parse this into the above-mentioned format:
{resume_txt}

"""
```

```python
def get_messages(prompt):
  messages = [
    {
      "role": "system",
      "content": "You are resume parser. Your task is to parse given resume into given format without excluding any details from the resume. Try to find a section where the info fits, else dump it seperately under others section."
    },
    {
    "role": "user",
    "content": prompt,
    }
  ]
  return messages

def get_completion(messages):
        response = openai.ChatCompletion.create(
            model=model,
            messages=messages,
            temperature=0
        )
        return response


(json.dumps(prompt))

result = get_completion(get_messages(prompt))
print(result.choices[0]['message']['content'])
```

```python
def get_messages_basic(prompt):
  messages = [
    {
    "role": "user",
    "content": prompt,
    }
  ]
  return messages
(json.dumps(basic_prompt))

result = get_completion(get_messages_basic(basic_prompt))
print(result.choices[0]['message']['content'])
```
